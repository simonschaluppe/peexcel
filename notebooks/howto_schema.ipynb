# Create schema folder

```python
# Function 1: Create Schema folder

from pexl.io.converter import xlsx_to_dataset_dir

version = "v1_12_7"

xlsx_to_dataset_dir(
    f"../data/exports/schema_{version}.xlsx", # created by PEExcel macro "export_schema"
    f"../data/schemas/{version}", #creates this directory and csv files
    mode="schema",
    replace_files=True # careful!
)

```

## Continue here
9. Check if folder was created in peexcel/data/schemas/vX_XX_X
10. Run Function 2: Show schema differences, which shows all differences between the old and new Schema

# Show differences in schema versions

```python
# Function 2: Show schema differences

from pexl.validate.schema import diff_schema_dirs, schema_diff_to_markdown
from IPython.display import display, Markdown

diff_result = diff_schema_dirs(
    old_dir=f"../data/schemas/v1_11_7",
    new_dir= f"../data/schemas/{version}"
)

md = schema_diff_to_markdown(diff_result, value_format="None")
display(Markdown(md))
```

11. If everything looks alright to you, run Function 3: Generate Schema Module
12. After that, run Function 4: Generate named variables
13. Run Schema test suite to verify
14. Success!

```python
# Function 3: Generate Schema Module

from pexl.schema.schema_codegen import generate_schema_module_from_csv_dir


generate_schema_module_from_csv_dir(
        schema_dir=f"../data/schemas/{version}",
        output_py_path=f"../src/pexl/schema/generated/excel_{version}.py",
        version=version,
    )
```

```python
# Function 4: Generate names variables

import pexl.schema.current as current_schema # latest schema must be manually set
current_schema.ExcelNamedVariables # is used by module excel 
```
