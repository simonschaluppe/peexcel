# Update python schema model
Only do this if you know what you are doing (so don't).

1. Increase Schema variant number in PEQExcel Input Sheet
2. Save File
3. Open VB Editor
4. Module a_export, Run Macro "Export Schema" by pressing F5
5. In the popup navigate to peexcel/data/exports, press save
6. Check if file is saved in peexcel/data/exports
7. Set correct version = version name in Schema folder code
8. Run Schema folder code
9. Check if folder was created in peexcel/data/schemas/vX_XX_X

# Create Schema folder

````python
# Schema folder code

from pexl.io.converter import xlsx_to_dataset_dir

version = "v1_11_5"

xlsx_to_dataset_dir(
    f"../data/exports/schema_{version}.xlsx", # created by PEExcel macro "export_schema"
    f"../data/schemas/{version}", #creates this directory and csv files
    mode="schema",
    replace_files=True # careful!
)

````

# Schema debugging

````python
from pexl.validate.schema import diff_schema_dirs

diff_result = diff_schema_dirs(
    old_dir=f"../data/schemas/v1.10_4",
    new_dir= f"../data/schemas/{version}"
)
diff_result
````

````python
from pexl.validate.schema import schema_diff_to_markdown


md = schema_diff_to_markdown(diff_result, value_format="None")
print(md)
````
